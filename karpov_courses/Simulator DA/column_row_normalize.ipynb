```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Column normalization (Z-Score Normalization)

```python
# Create a dataframe where the average value of the second column is higher than others:
df = pd.DataFrame(np.random.randn(10,10)* 4 + 3)
df[1] = df[1]+40
df
```

```python
# If we do a heatmap, we just observe that one column has higher values than others:
sns.heatmap(df, cmap='viridis')
plt.show()
```

```python
# Now if we normalize it by column (Z-Score Normalization):
df_norm_col = (df - df.mean())/df.std()
df_norm_col
```

```python
sns.heatmap(df_norm_col, cmap='viridis')
plt.show()
```

# Row normalization

```python
# Create a dataframe where the average value of the second row is higher
df_2 = pd.DataFrame(np.random.randn(10,10) * 4 + 3)
df_2.iloc[2]=df.iloc[2]+40
df_2
```

```python
# If we do a heatmap, we just observe that one row has higher values than others:
sns.heatmap(df_2, cmap='viridis')
plt.show()
```

```python
# Normalize it by row:
df_norm_row = df_2.apply(lambda x: (x-x.mean())/x.std(), axis = 1)
```

```python
# And see the result
sns.heatmap(df_norm_row, cmap='viridis')
plt.show()
```

# Feature	Normalization (Min-Max Scaling)	

**Purpose**:	Scales data to a fixed range (usually [0, 1])   
**Formula**:	X_new = (X - X_min) / (X_max - X_min).  
**Resulting Range**:	Bounded (e.g., 0 to 1).  
**Affected by Outliers?**	Yes, very sensitive. A single outlier changes min/max.  
**Use Case**: Algorithms that require data on a consistent scale (e.g., images, neural networks, K-NN).
